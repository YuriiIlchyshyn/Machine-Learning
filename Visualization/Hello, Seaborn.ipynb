## Set up the notebook
Спочатку налаштуємо середовище і додамо потрібні бібліотеки

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Load the data
У цьому notebook ми будемо працювати з набором даних про історичні рейтинги FIFA для шести країн: Аргентина (ARG), Бразилія (BRA), Іспанія (ESP), Франція (FRA), Німеччина (GER) та Італія (ITA). Набір даних зберігається у вигляді файлу CSV (короткий для файлу значень, розділених комами. Відкриття файлу CSV у Excel показує рядок для кожної дати разом із стовпцем для кожної країни.

Щоб завантажити дані в зошит, ми скористаємося двома кроками:
* почнемо із вказівки місця (або файлового шляху), де можна отримати доступ до набору даних, а потім
* використаємо шлях до файлу, щоб завантажити вміст набору даних у notebook

```python
# Path of the file to read
fifa_filepath = "fifa.csv"

# Read the file into a variable fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

### Comments
Два рядки після знаку решітки (`#`) називають коментарями.
Обидва рядки повністю ігноруються комп'ютером при запуску коду, і вони потрібні для того, щоб людина, яка читає код, може швидко зрозуміти його.
### Executable code
Інші два рядки - це код, який раниться комп'ютером (у цьому випадку для пошуку та завантаження набору даних).
Перший рядок присвоює змінній `fifa_filepath` локацію місця, де можна отримати доступ до набору даних. У цьому випадку ми присвоїли. Зауважимо, що коментар безпосередньо над цим рядком виконуваного коду дає короткий опис того, що він робить!
Другий рядок - змінна `fifa_data`, що містить всю інформацію в наборі даних. Це робиться за допомогою `pd.read_csv`. 

Ці деталі матимуть більший сенс незабаром, коли ти матимеш можливість завантажити власний набір даних у практичній роботі.
Поки що важливо пам’ятати, що кінцевим результатом запуску обох рядків коду є те, що тепер ми можемо отримати доступ до набору даних із notebook за допомогою fifa_data.

## Examine the data
Тепер ми швидко розглянемо набір даних у `fifa_data`, щоб переконатися, що він завантажений належним чином.
Друкуємо перші п'ять рядків набору даних, записуючи один рядок коду так:
* почнемо зі змінної, що містить набір даних (у цьому випадку `fifa_data`), а потім

```python
# Print the first 5 rows of the data
fifa_data.head()
```

## Plot the data
Ми розглянемо багато видів графіків. В більшості випадків графік можна намалювати, написавши всього один рядок коду!

Для початку намалюємо лінійну діаграму (line chart)

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```
