# Візуалізація даних

### Що таке дані і для чого вони потрібні?
[Дані](https://uk.wikipedia.org/wiki/%D0%94%D0%B0%D0%BD%D1%96) простими словами - це набір інформації, на основі якого ми можемо будувати певні висновки, рішення і тд.
Чим краще ми опрацьовуємо дані, тим більшу користь з тієї інформації ми отримуємо.
Одним зі способів подання даних - це візуалізація.

### Візуаізація даних
Візуалізація даних - це велика частина роботи фахівців [data science](https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%83%D0%BA%D0%B0_%D0%BF%D1%80%D0%BE_%D0%B4%D0%B0%D0%BD%D1%96). На ранніх стадіях розвитку проекту часто необхідно виконувати розвідувальний аналіз даних (Exploratory data analysis (EDA)) для виявлення закономірностей у наборі даних. Візуалізація даних допомагає уявити великі і складні набори даних в простому вигляді. На етапі закінчення проекту важливо зуміти відзвітувати про його результати так, щоб навіть людям, які не володіють технічними знаннями, все було зрозумілим.

### Способи візуалізації в Python:
* 1 спосіб: Руцями. Тобто ти сам пишеш функції, вигадуєш алгоритми і тд 
* 2 спосіб: За допомогою готових бібліотек.
Очевидно, що незважаючи на "креативність" 1 способу він є неефективним, затратним та недосконалим.
2 спосіб є практичнішим та швидшим, тому сьогодні ми розглянемо саме його

## Найпопулярніші бібліотеки для візуалізації даний в Python

## Matplotlib - Mister Oldschool

Вперше світ почув про Matplotlib у 2003 році. Бібліотека Matplotlib побудована на принципах ООП, але має процедурний інтерфейс pylab, який надає аналоги команд MATLAB.
Пакет підтримує багато видів графіків і діаграм:

* line plot
* scatter plot
* bar chart
* histogram
* pie chart
* stem plot
* contour plot
* quiver
* spectrogram

Оскільки Matplotlib був першою бібліотекою візуалізації даних Python, багато інших бібліотек будуються над ним або розроблені для роботи в тандемі з ним під час аналізу. Деякі бібліотеки, такі як Pandas та Seaborn, є «обгортками» над Matplotlib. Вони дозволяють отримати доступ до ряду методів Matplotlib з меншим кодом.

Хоча Matplotlib корисний для розуміння даних, він не дуже корисний для швидкого та легкого створення графіків якості публікацій. Як вказує Кріс Моффіт у своєму огляді інструментів візуалізації Python, Matplotlib "надзвичайно потужний, але з цією силою виникає складність".

Matplotlib давно критикується за стилі за замовчуванням, які мають чітке відчуття 1990-х.

Плюси і мінуси: 
Matplotlib досі не втрачає своїх позицій і залишається в топі найпопулярніших інструментів для візуалізації інформації. Але основною причиною такої "стабільності" є те, що людям складно відмовлятися від того, що вони довго вчили та звикли. Якщо ти хочеш повноцінно працювати з Matplotlib, то твій найкращий друг - StackOverflow. Як не крути, зараз існує багато альтернатив та аналогів. які мають простіши синтаксис та приємніші у використанні. Важливо зазначити, що Matplotlib таки не розрахований на величезні об'єми даних, тому всілякі шаманські конструкції так складно писати.

Корисні лінки:

## Seaborn -  my little clone

Seaborn використовує силу matplotlib для створення прекрасних діаграм у кількох рядках коду. Ключова відмінність - стилі та кольорові палітри Seaborn за замовчуванням, які розроблені таким чином, щоб бути більш естетично приємними та сучасними. Оскільки Seaborn побудований на вершині matplotlib, вам потрібно знати matplotlib, щоб змінити типові значення Seaborn.

Корисні лінки:
* https://seaborn.pydata.org/introduction.html
* https://seaborn.pydata.org/tutorial.html

## Ggplot - little brother Ggplot2 from planet R

Ggplot ґрунтується на Ggplot2, R-графічній системі та [The Grammar of Graphics](https://www.amazon.com/Grammar-Graphics-Statistics-Computing/dp/0387245448). Ggplot працює інакше, ніж Matplotlib: він дозволяє компонентам шару створювати повноцінний сюжет. Наприклад, ви можете починати з осей, потім додавати точки, потім лінію, лінію тренду і т.д. The Grammar of Graphics оцінили як "інтуїтивний" метод побудови графіків, досвідченим користувачам Matplotlib може знадобитися час, щоб пристосуватися до цього нового мислення.

За словами автора, Ggplot не розроблений для створення високомодифікованої графіки. Він жертвує складністю для більш простого способу побудови графіків.

Ggplot тісно інтегрований з Pandas, тому найкраще зберігати свої дані в DataFrame при використанні Ggplot.

Корисні лінки:
* http://ggplot.yhathq.com/
* http://ggplot.yhathq.com/how-it-works.html

## Bokeh - little cousine Ggplot from planet Python

Як і Ggplot, Bokeh базується на The Grammar of Graphics, проте, бібліотека Bokeh є рідною для Python. Її сила полягає у здатності створювати інтерактивні, web-ready графіки, які можна легко виводити як об’єкти JSON , HTML-документи або інтерактивні вебаплікації. Bokeh також підтримує потокову передачу даних у режимі реального часу.

Bokeh пропонує три інтерфейси з різними рівнями управління для розміщення різних типів користувачів.
Найвищий рівень - це швидке створення діаграм. Він включає способи створення загальних діаграм, таких як діаграми, графіки вікон та гістограми. 
Середній рівень має ту саму специфіку, що і matplotlib, і дозволяє керувати основними будівельними блоками кожної діаграми (наприклад, точки на графіці розкидання). 
Найнижчий рівень орієнтований на розробників та програмних інженерів. У ньому немає попередньо встановлених значень за замовчуванням і вимагає визначити кожен елемент діаграми.

Корисні лінки:
* https://docs.bokeh.org/en/latest/index.html
* https://mybinder.org/v2/gh/bokeh/bokeh-notebooks/master?filepath=tutorial%2F00%20-%20Introduction%20and%20Setup.ipynb

## Plotly or "take it easy"

[Plotly](https://plotly.com/) - це графічна компанія з кількома продуктами і open-source інструментами.
Перший реліз plotly.py Graphing Library відбувся у 2013 році. 
Сьогодні існують 3 Graphing Library:

* Plotly Python Open Source Graphing Library
* Plotly R Open Source Graphing Library
* Plotly JavaScript Open Source Graphing Library

Пакет plotly для Python - це open-source бібліотека, заснована на [plotly.js](https://plotly.com/javascript/), яка, в свою чергу, заснована на [d3.js](https://d3js.org/).

Plotly відомий як онлайн-платформа для візуалізації даних. Як і в Bokeh, в Plotly можна створити інтерактивні графіки, та в його арсеналі є деякі діаграми, яких не знайти в більшості бібліотек, як контурні графіки, дендограми та 3D-діаграми.
Плюси і мінуси:
Plotly крута бібліотека для роботи з даними. Першою та очевидною перевагою є те, що вона open-source. Графіки зазвичай малюються в одну стрічку коду за короткий термін часу. Проте, як на мене, найкращим є те, що всі графіки є інтерактивними і їх можна переглядати на NBViewer. З мінусів є те. що зараз існує не настільки велика кількість тутуоріалів, як для Matprolib. Проте ситуацію рятує класна документація

Корисні лінки: 
* https://plotly.com/python/
* https://habr.com/ru/post/308162/
* https://tproger.ru/translations/python-data-vizualization/
* https://nbviewer.jupyter.org/github/WillKoehrsen/Data-Analysis/blob/master/plotly/Plotly%20Whirlwind%20Introduction.ipynb

## Pygal - best friend for small dataset, which like Rock

Як і Bokeh та Plotly, Pygal малює інтерактивні графіки, які можна вбудувати у веб-браузер. Її головним диференціатором є можливість виведення графіків у вигляді SVG. АЛЕ: Поки ви працюєте з меншими наборами даних, SVG будуть чудесними. Але якщо ви створюєте діаграми з сотнями тисяч даних, у них виникнуть проблеми з рендерінгом і SVG стануть млявими.

Оскільки кожен тип діаграми упакований у метод, а вбудовані стилі досить гарні, легко створити красиву діаграму, написавши лише кілька рядків коду

Корисні лінки:
* http://www.pygal.org/en/latest/index.html
* http://www.pygal.org/en/latest/documentation/index.html

## Geoplotlib -  girlfriend from the Faculty of Geography

Geoplotlib - це інструментарій для створення карт та побудови географічних даних. Ви можете використовувати його для створення різних типів карт, таких як хороплети, теплові карти та карти густини точок. Для використання Geoplotlib у вас повинен бути встановлений Pyglet. Тим не менше, оскільки більшість бібліотек візуалізації даних Python не пропонують карт, приємно мати бібліотеку, присвячену лише їм.
