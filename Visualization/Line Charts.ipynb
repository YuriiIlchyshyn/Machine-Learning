## Plot the data
Тепер, коли ми успішно зробили попередні кроки можемо малювати графіки

```python
# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa)
```

Як бачимо, рядок коду є відносно коротким і має дві основні складові:
* `sns.lineplot` повідомляє notebook, що ми хочемо створити лінійну діаграму.
> Кожна команда, яка починається з `sns`, що вказує на те, що вона належить бібліотеці **seaborn**. Наприклад, ми використовуємо `sns.lineplot` для створення лінійних діаграм. Незабаром ми розглянемо `sns.barplot` та `sns.heatmap`, які викорристовуються для створення графіків та теплових карт відповідно.
* `data=fifa` вибирає дані, які будуть використані для створення діаграми.

Іноді є певні деталі та особливості, які ми хотіли б змінити, наприклад розмір фігури та заголовок діаграми. Кожен з цих параметрів легко додається одним рядком коду.

```python
# Set the width and height of the figure
plt.figure(figsize=(12,6))

# Add title
plt.title("Fifa score")

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa)
```

Перший рядок коду встановлює розмір фігури: 16 дюймів (в ширину) на 6 дюймів (у висоту). Якщо ми хочемо використовувати власний розмір, змінюємо передбачені значення 16 та 6 на потрібну ширину та висоту.

Другий рядок коду задає заголовок фігури. **Зауваження: заголовок завжди повинен бути у лапках ("...")!**

У наступній code cell ми побудуємо рядки, що відповідають першим двом стовпцям набору даних.

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Historical FIFA rankings")

# Line chart showing how FIFA rankings evolved over time for 'ARG'
sns.lineplot(data=fifa['ARG'], label="Argentina")

# Line chart showing how FIFA rankings evolved over time for 'BRA'
sns.lineplot(data=fifa['BRA'], label="Brazilia")

# Add label for horizontal axis
plt.xlabel("Date")
plt.ylabel("Score")
```

Перші два рядки коду задають назву та розмір фігури (і повинні виглядати дуже знайомо).
