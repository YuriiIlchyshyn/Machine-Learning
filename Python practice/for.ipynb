### Розглянемо наступний приклад

```python
hugehairypants = ['величезні', 'махнаті', 'штани']
for i in hugehairypants:
    print(i)
    print(i)
```

#### ВАЖИЛИВО: СЛІДКУЙ ЗА ПРОБІЛАМИ
Цей блок коду не спрацює через зайвий пробіл в останній стрічці

```python
hugehairypants = ['величезні', 'махнаті', 'штани']
for i in hugehairypants:
    print(i)
     print(i)
```

#### Уважно проаналізуй наступний приклад

```python
hugehairypants = ['величезні', 'махнаті', 'штани']
for i in hugehairypants:
    print("Цикл верхній номер кроку" , i)
    for j in hugehairypants:
        print("Цикл всередині номер кроку", j)
    print("---------------")        
```

### Згадаємо приклад про монети

Порахуємо скільки монет ми матимемо у кінці року враховуючи, що щотижня ворона краде 3 монети

```python
found_coins = 20
magic_coins = 70
stolen_coins = 3
coins = found_coins
for week in range(1, 53):
    coins = coins + magic_coins - stolen_coins
    print('Тиждень %s = %s' % (week, coins))
```
