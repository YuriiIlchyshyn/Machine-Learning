### На касі йому запропонували взяти пакет, проте він не знав якого розміру його покупки, тому:

### Потім наш друг вирішив піти в Сільпо і купити щось екзотичне

```python
exotic_list = ["apple", "banana", "cherry","orange","mango"]
```

### Він видалив банани, бо не любив їх через колір

```python
exotic_list.remove("banana")
print(exotic_list)
```

### І перший елемент в списку

```python
del exotic_list[0]
print(exotic_list)
```

### А ще він фанат pop музики, і взнавши що в пайтоні є така функція, вирішив її використати

```python
thislist = ["apple", "banana", "cherry","orange","mango",'pear','grape']
print(thislist)
thislist.pop()# removes the last item if index is not specified
print(thislist)
thislist.pop(0)
print(thislist)
```

### Йому набридло гратися зі списками, тому він спершу видалив 2 елемент, тобто 'cherry'

```python
del thislist[1]
print(thislist)
```
