```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

## Select a dataset
Цього разу ми  будемо працювати з набором даних Міністерства транспорту США (US Department of Transportation), який відстежує затримки рейсів.

Відкриваючи цей CSV-файл у Excel, у нас є місяць (де 1 = січень, 2 = лютий тощо) та код авіакомпанії.

Кожен запис показує середню затримку прильоту (у хвилинах) для різних авіакомпаній та місяць (усі у 2015 році). Негативні записи позначають рейси, які (в середньому) мали тенденцію прибувати рано. Наприклад, середній рейс American Airlines (код авіакомпанії: AA) в січні прибув приблизно на 7 хвилин пізніше, а рейс Alaska Airlines (код авіакомпанії: AS) у травні прибув приблизно на 3 хвилини раніше.

## Load the data
Як і раніше, ми завантажуємо наш датасет за допомогою команди pd.read_csv

```python
# Path of the file to read
flight_filepath = "flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col="Month")
```

Можна помітити, що код трохи коротший, ніж ми використовували в попередніх notebook. У цьому випадку, оскільки мітки рядків (у стовпці `'Month'`) не відповідають датам, ми не додаємо `parse_dates = True` в круглі дужки.
Але ми зберігаємо перші два фрагменти тексту, як і раніше, щоб забезпечити:
* файловий шлях для набору даних (в даному випадку, `file_filepath`) та
* назва стовпця, який буде використовуватися для індексації рядків (у цьому випадку `index_col = "Month"`).

## Bar chart
Припустимо, що ми хочемо створити діаграму із середньою затримкою прильоту для рейсів Spirit Airlines (код авіакомпанії: NK) за місяцем.

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.цex, y=flight_data['UA'])

# Add label for vertical axis
plt.ylabel("Arrival delay (in minutes)")
```

Команди для налаштування тексту (тобто заголовків та осей) і розміру нашої фігури відомі ще з попкркдніх notebook.
Тому розглянемо лише код який створює bar chart:

`# Bar chart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])`
Тут є три основних компоненти:
* `sns.barplot` - повідомляє, що ми хочемо створити bar chart
* `x = flying_data.index` - визначає, що використовувати на горизонтальній осі. У цьому випадку ми вибрали стовпець, який індексує рядки (тобто стовпець, який містить місяці)
* `y = fly_data ['NK']` - встановлює, який стовпець в даних, буде використовуватися для визначення висоти кожного стовпця. У цьому випадку ми вибираємо стовпчик 'NK'.

## Heatmap
У нас залишився  ще один вид графіків, які ми вивчатимемо в цьому notebook - Heatmap
В наступному code cell ми створимо heatmap для швидкоъ вызуалызацыъ нашого `flight_data`. Тут кожен колір відповідає певному значенню. Чим темніший колір, тим менше значення. Значення NaN зафарбовуються у білий колір.

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add title
plt.title("Average Arrival Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add label for horizontal axis
plt.xlabel("Airline")
```

Розглянемо наступний уривок коду:

`# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)`
Він містить три основних компоненти:
* `sns.heatmap` - повідомляє, що ми хочемо створити heatmap
* `data=flight_data` - вказує, що ми хочемо використовувати всі записи в `flight_data` для створення heatmap
* `annot=True` - гарантує, що значення для кожної комірки відображаються на графіку
